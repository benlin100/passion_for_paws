### creating binary columns for each color

```python
import re
re.compile('<title>(.*)</title>')

def extract_unique_colors(train):
    """
    Extract a set of unique colors from the training set with
    100 or more animals.
    """
    colors = {}
    for color in list(train['color']):
        color = re.split('\W+', color)
        for subcolor in color:
            try:
                colors[subcolor] += 1
            except:
                colors[subcolor] = 1
    return set([k for k, v in colors.items() if v >= 0])


def create_color_variables(data, colors):
    vectors = []
    to_position = dict([(c, i) for i, c in enumerate(colors)])
    for color in list(data['color']):
        vector = [0] * len(colors)
        color = re.split('\W+', color)
        for subcolor in color:
            try:
                vector[to_position[subcolor]] += 1
            except:
                pass
        vectors.append(vector)
    columns = ['color_' + x[1] for x in sorted(
            [(v, k) for k, v in to_position.items()])]
    colors_df = pd.DataFrame(vectors, columns = columns)
    data.reset_index(inplace=True, drop=True)
    colors_df.reset_index(inplace=True, drop=True)
    result = pd.concat([data, colors_df], axis=1, join_axes=[data.index]) 
    result.reset_index(inplace=True, drop=True)
    return result
```

```python
colors = extract_unique_colors(animal_data)
animal_data = animal_data.pipe(create_color_variables, colors = colors)
```

```python
import re
re.compile('<title>(.*)</title>')

def extract_unique_colors(train):
    """
    Extract a set of unique colors from the training set with
    100 or more animals.
    """
    colors = {}
    for color in list(train['breed']):
        color = re.split('\/+', color)
        for subcolor in color:
            try:
                colors[subcolor] += 1
            except:
                colors[subcolor] = 1
    return set([k for k, v in colors.items() if v >= 0])

def create_color_variables(data, colors):
    vectors = []
    to_position = dict([(c, i) for i, c in enumerate(colors)])
    for color in list(data['breed']):
        vector = [0] * len(colors)
        color = re.split('\/+', color)
        for subcolor in color:
            try:
                vector[to_position[subcolor]] += 1
            except:
                pass
        vectors.append(vector)
    columns = ['breed_' + x[1] for x in sorted(
            [(v, k) for k, v in to_position.items()])]
    colors_df = pd.DataFrame(vectors, columns = columns)
    result = pd.concat([data, colors_df], axis=1, join_axes=[data.index]) 
    #result.reset_index(inplace=True, drop=True)
    return result

colors = extract_unique_colors(animal_data)
animal_data = animal_data.pipe(create_color_variables, colors = colors)
```
